```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u

import sys
sys.path.insert(0, '../')
```

```python
from libra import IRTFTemplate

star = IRTFTemplate('K2V')
spots = star.scale_temperature(-200)

star.plot(label='photosphere')
spots.plot(label='spots')

# Fractional spotted area
f_s = 0.1

combined = (1 - f_s) * star + f_s * spots

combined.plot(label='combination')

plt.legend()
```
