```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u

import sys
sys.path.insert(0, '../')
```

```python
from libra import IRTFTemplate

star = IRTFTemplate('M8V')
spots = star.scale_temperature(-200)

star.plot(label='photosphere', ls='--', color='DodgerBlue', alpha=0.5)
spots.plot(label='spots', ls="--", color='r', alpha=0.5)

# Fractional spotted area
f_s = 0.5

combined = (1 - f_s) * star + f_s * spots

combined.plot(label='combination', color='k')

plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Flux [{0}]'.format(combined.flux.unit.to_string()))

plt.legend()
plt.show()
```
